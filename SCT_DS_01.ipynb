```python
import pandas as pd
import matplotlib.pyplot as plt

# Load World Bank population dataset
data = pd.read_csv(r"C:\Users\Lenovo\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_40826\API_SP.POP.TOTL_DS2_en_csv_v2_40826.csv", skiprows=4)

# Check the first few rows
print(data.head())
```

```python
# Select countries and year
countries = ["India", "China", "United States"]
year = "2022"

# Filter data
filtered_data = data[data["Country Name"].isin(countries)]

# Create bar chart
plt.figure()
plt.bar(filtered_data["Country Name"], filtered_data[year])
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Population Comparison by Country (2022)")
plt.show()
```
